```python
#Simple function
def greet():
    print("Hello World")

greet()
#Function with parameter
def greet(name):
    print("Hello", name)

greet("Samruddhi")
```

```python
#Function to add numbers
def add(a, b):
    result = a + b
    print("Sum =", result)

add(5, 7)
```

```python
#Function with return value
def multiply(a, b):
    return a * b

result = multiply(4, 5)

print(result)
```

```python
#Function with loops
def print_numbers():
    for i in range(1, 6):
        print(i)

print_numbers()
```

```python
#Even or Odd function
def check_even_odd(num):
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")

check_even_odd(10)
check_even_odd(11)
```
