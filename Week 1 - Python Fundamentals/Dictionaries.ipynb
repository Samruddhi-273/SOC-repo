```python
#Dictionaries
student = {
    "Name": "Samruddhi",
    "Age": 20,
    "Branch": "Aerospace"
}
student["City"] = "Mumbai"

print(student)
print(student["City"])
```

```python
#Loop through dictionaries
student = {
    "Name": "Samruddhi",
    "Age": 20,
    "Branch": "Aerospace",
    "City": "Mumbai",
}

for key, value in student.items():
    print(key, ":", value)
```
