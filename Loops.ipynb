```python
#for loop
for i in range(5):
  print(i)
  #print 1 to 10
for i in range(1, 11):
    print(i)
    #print even numbers
for i in range(2, 11, 2):
    print(i)
```

```python
#while loop
count = 1

while count <= 5:
    print(count)
    count += 1
```

```python
#loop through lists
fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)

names = ["iitb", "powai"]
for name in names:
  print(name)
```

```python
#Pattern problem
for i in range(6):
    print("*" * i)
```
