```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame({'area':[2600,3000,3200,3600,4000],'price':[550000,565000,610000,680000,725000]
                   })
print(df)
```

```python
plt.scatter(df['area'], df['price'])
plt.xlabel("Area (sq ft)")
plt.ylabel("Price (Lakh ₹)")
plt.title("House Price Dataset")
plt.show()
```

```python
model = LinearRegression()

model.fit(df[['area']], df['price'])
```

```python
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
prediction = model.predict([[3300]])

print("Predicted Price:", prediction[0])
```

```python
areas = [[1200], [1800], [2700], [3500]]

predictions = model.predict(areas)

for area, price in zip(areas, predictions):
    print(f"Area: {area[0]} sq ft -> Price: {price:.2f} lakh")
```

```python
plt.scatter(df['area'], df['price'])

plt.plot(
    df['area'],
    model.predict(df[['area']])
)

plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Linear Regression Best Fit Line")

plt.show()
```
